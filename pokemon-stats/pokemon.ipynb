# Pokemon Data Set
This data set includes all pokemon currently in the game. It includes their name, types, combat stats, generation, and Legendary status.
<br>
This notebook will explore this data with graphs and draw conclusions based on findings.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
# Show the first 5 rows of the data set.
# Show the info of the data set.
df = pd.read_csv('./Pokemon.csv')
df
```

```python
x = df['Attack'].values
y = df['Defense'].values
```

```python
X = x.reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(X,y,train_size=0.8, test_size=0.2, random_state=100)
```

```python
%matplotlib inline
plt.scatter(x_train, y_train, color='purple')
plt.xlabel('Attack Value')
plt.ylabel('Defense Value')
plt.title('Pokemon Power Level')
plt.show()
```

```python

lm = LinearRegression()
lm.fit(x_train, y_train)
y_predict = lm.predict(x_test)
print(f"Train Accuracy {round(lm.score(x_train, y_train)*100,2)}%")
print(f"Test Accuracy {round(lm.score(x_test, y_test)*100,2)}%")
```

```python
plt.scatter(x_train, y_train, color = 'Purple')
plt.plot(x_test, y_predict)
plt.xlabel("Attack Value")
plt.ylabel("Defense Value")
plt.title('Pokemon Power Level')
plt.plot
```
